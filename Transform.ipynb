# Data Prep Apple-Price-List

```python
df_apple = pd.read_csv("./apple product price list from 26 countries.csv")
df_apple['country_code']=df_apple['country_code'].str.upper()
df_apple.head()
```

```python
df_apple['model_en']=df_apple['model']

df_apple.loc[df_apple['model'].str.contains('imac', case=False), 'model_en'] = '24-inch iMac'
df_apple.loc[df_apple['model'].str.contains('pen', case=False), 'model_en'] = 'Apple Pencil (2nd generation)'
df_apple.loc[df_apple['model'].str.contains('spor|Спортивный ремешок|Correa deportiva|Urheiluranneke', case=False), 'model_en'] = 'Sport Band'
df_apple.loc[df_apple['model'].str.contains('airpods\(2|airpods\(andra|airpods\(รุ่นที่ 2|airpods\(seconda|airpods\(segunda', case=False), 'model_en'] = 'AirPods(2nd generation)'
df_apple.loc[df_apple['model'].str.contains('airpods\(3|airpods\(รุ่นที่ 3|airpods\(tre|airpods\(ter', case=False), 'model_en'] = 'AirPods(3rd generation)'
df_apple.loc[df_apple['model'].str.contains('mouse', case=False), 'model_en'] = 'Magic Mouse'
df_apple.loc[df_apple['model'].str.contains('Apple Store|iTunes', case=False), 'model_en'] = 'App Store and iTunes Gift Card'
```

Bereinigen von lückenhaften Werten

```python
df_apple = df_apple.drop(df_apple[df_apple.model_en == "iPhone 13 Pro"].index)
df_apple = df_apple.drop(df_apple[df_apple.model_en == "iPhone 13"].index)
df_apple = df_apple.drop(df_apple[df_apple.model_en == "Apple TV HD"].index)
df_apple = df_apple.drop(df_apple[df_apple.model_en == "App Store and iTunes Gift Card"].index)
df_apple = df_apple.drop(df_apple[df_apple.model_en == "AirPods(3rd generation)"].index)
df_apple = df_apple.drop(df_apple[df_apple.model_en == "AirPods Max"].index)
df_apple.head()
```

Bereinigung aller Textzeichen --> Nur noch Zahlen + Komma/Punkte

```python
df_apple['price2']=df_apple['price'].str.split('or', 1, expand=True)[0]
df_apple['price2']=df_apple['price2'].str.replace('[^\d\.\,]', '', regex=True)
df_apple.head()
```

Harmonisierung der Kommas/Punkte

```python
comma_decimal_point_locale = ['AM', 'AR', 'AT', 'AZ', 'BA', 'BE', 'BG', 'BO', 'BR', 'BY',
                    'CA', 'CH', 'CL', 'CM', 'CO', 'CR', 'CY', 'CZ', 'DE', 'DK',
                    'EC', 'EE', 'ES', 'FI', 'FO', 'FR', 'GE', 'GL',  'HR', 'HU',
                    'ID', 'IS', 'IT', 'KZ', 'LB', 'LT', 'LU', 'LV', 'MA', 'MD', 'MK', 'MO', 'MZ',
                    'NL', 'NO', 'PE', 'PL', 'PT', 'PY', 'RO', 'RS', 'RU',
                    'SE', 'SI', 'SK', 'TN', 'TR', 'UA', 'UY', 'UZ', 'VE', 'VN', 'ZA']

price_clean=[]
for i, r in df_apple.iterrows():
    p=r['price2']
    if r['country_code'] in comma_decimal_point_locale:
        p=p.replace('.', '').replace(',', '.')
    else:
        p=p.replace(',', '')
    price_clean.append(p)
    
df_apple['price_local_currency']=price_clean
df_apple['price_local_currency']=pd.to_numeric(df_apple['price_local_currency'])
df_apple.head()
```
